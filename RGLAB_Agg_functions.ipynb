```python
import pandas as pd 
import numpy as np
```

```python
data = {
    'Category': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Value': [10, 15, 20, 25, 30, 35],
    'Fee' :[20000,25000,26000,22000,24000,35000],
    'Duration':['30day','40days','35days','40days','60days','60days'],
    'Discount':[1000,2300,1200,2500,2000,2000]
}

df = pd.DataFrame(data)
```

```python
df.head(6)
```

```python
# apply single aggregate func
total_sum = df['Value'].aggregate('sum')
print('Total sum:', total_sum)
```

```python
sum2 = df['Value'].sum()
print('Totalsum2:', sum2)
```

```python
# cumsum() running total
df[['Value']].cumsum()
```

```python
# mean/avg
t_mean = df['Value'].agg('mean')
print('Avg mean:', t_mean)

#alternative
mean = df['Value'].mean()
print(f'Mean: {mean}')
```

```python
max_v = df['Value'].agg('max')
print('Max Value:', max_v)
```

```python
dev = df['Value'].agg('count')
print('Count:', dev)

count = df['Value'].count()
print('Count',count)
```

```python
std = df['Value'].agg('std')
print('STd:', std)

std1 = df['Value'].std()
print('STD1:', std1)
```

```python
result = df[['Fee', 'Discount']].agg('sum')

result
```

```python
# determine the total of each group an agg func

result_g = df.groupby('Category')[['Fee', 'Discount']].agg('sum')
result_g
```
